```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

dataset = pd.read_csv('./Market_Basket_Optimisation.csv',header=None)
print(dataset.shape)


# 基于mlxtend的数据关联分析
transactions = []
for i in range(0,dataset.shape[0]):
    temp=[]
    for j in range(0,20):
        # print(dataset.values[i,j])
        if str(dataset.values[i,j]) !='nan':
            temp.append(str(dataset.values[i,j]))
    transactions.append(temp)
    
#对数据进行one-hot编码    
temp2 = TransactionEncoder()
temp2_hot_encoded = temp2.fit(transactions).transform(transactions)
df = pd.DataFrame(temp2_hot_encoded, columns=temp2.columns_)


# 挖掘频繁项集
itemsets = apriori(df,use_colnames=True, min_support=0.05)
itemsets = itemsets.sort_values(by="support" , ascending=False) 
print('-'*20, '频繁项集', '-'*20)
print(itemsets)

# 根据频繁项集计算关联规则
rules =  association_rules(itemsets, metric='lift', min_threshold=1)
rules = rules.sort_values(by="lift" , ascending=False) 
print('-'*20, '关联规则', '-'*20)
print(rules)
```
