# 汽车质量数据统计

```python
# Step1，数据加载
df_complaint=pd.read_csv('car_complain.csv')
```

```python
# Step2，数据预处理 拆分problem类型 
df_complaint.drop('problem',axis=1).join(df_complaint.problem.str.get_dummies(','))
```

```python
# 品牌投诉总数
df_brand=df_complaint['brand'].value_counts()
df_brand
```

```python
# 车型投诉总数
df_complaint['car_model'].value_counts()
```

```python
# 品牌车型数
df_model=df_complaint.groupby('brand')['car_model'].nunique()
df_model
```

```python
# 品牌平均车型投诉数
df_brand.sort_index()/df_model.sort_index()
```
